```python
distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": None},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": None}
}
distribuicoes
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = np.random.random(size=1000)
dados_norm = np.random.normal(0,1,size=1000)
dados_exp = np.random.exponential(2,size=1000)
```

```python
distribuicoes["Normal"]['dados'] = dados_norm
distribuicoes["Uniforme"]['dados'] = dados
distribuicoes["Exponencial"]['dados'] = dados_exp

distribuicoes['Normal']['estatisticas'] = {'média': np.mean(dados_norm), 'mediana':np.median(dados_norm),'variancia': np.var(dados_norm), 'desvio-padrao': np.std(dados_norm)}
distribuicoes['Uniforme']['estatisticas'] = {'média': np.mean(dados), 'mediana':np.median(dados), 'variancia':np.var(dados), 'desvio-padrao':np.std(dados)}
distribuicoes['Exponencial']['estatisticas'] = {'média': np.mean(dados_exp), 'mediana':np.median(dados_exp), 'variancia':np.var(dados_exp), 'desvio-padrao':np.std(dados_exp)}

for k in distribuicoes.keys():
    print(distribuicoes[k]['estatisticas'])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].hist(dados)
axs[1].hist(dados_norm)
axs[2].hist(dados_exp)
```

```python
estats = {}

for k in distribuicoes.keys():
  estats[k]  = {**distribuicoes[k]['estatisticas']}


estats_df = pd.DataFrame(estats)
estats_df
```

```python
distribuicoes['Poisson'] = {"params": {"lam": 2}, "dados": None, "estatisticas": None}
distribuicoes['Binomial'] = {"params": {"n": 1000, "p": 0.5}, "dados": None, "estatisticas": None}



dados_poiss = np.random.poisson(2,size=1000)
dados_bin = np.random.binomial(1000,0.5,size=1000)

distribuicoes["Poisson"]['dados'] = dados_poiss
distribuicoes["Binomial"]['dados'] = dados_bin

distribuicoes['Poisson']['estatisticas'] = {'média': np.mean(dados_poiss), 'mediana':np.median(dados_poiss),'variancia': np.var(dados_poiss), 'desvio-padrao': np.std(dados_poiss)}
distribuicoes['Binomial']['estatisticas'] = {'média': np.mean(dados_bin), 'mediana':np.median(dados_bin), 'variancia':np.var(dados_bin), 'desvio-padrao':np.std(dados_bin)}


for k in distribuicoes.keys():
    print(distribuicoes[k]['estatisticas'])
```

```python
estats = {}

for k in distribuicoes.keys():
  estats[k]  = {**distribuicoes[k]['estatisticas']}


estats_df = pd.DataFrame(estats)
estats_df
```

```python
for k in distribuicoes.keys():
  distribuicoes[k]['estatisticas']['min'] = np.min(distribuicoes[k]['dados'])
  distribuicoes[k]['estatisticas']['max'] = np.max(distribuicoes[k]['dados'])
  distribuicoes[k]['estatisticas']['q1'] = np.percentile(distribuicoes[k]['dados'],25)
  distribuicoes[k]['estatisticas']['q3'] = np.percentile(distribuicoes[k]['dados'],75)
```

```python
estats = {}

for k in distribuicoes.keys():
  estats[k]  = {**distribuicoes[k]['estatisticas']}


estats_df = pd.DataFrame(estats)
estats_df
```

```python
fig, axs = plt.subplots(1, 5, figsize=(18, 5))

axs[0].boxplot(dados_norm)
axs[0].set_title("Normal")
axs[0].set_xlabel("")
axs[0].set_ylabel("Valores")

axs[1].boxplot(dados)
axs[1].set_title("Uniforme")

axs[2].boxplot(dados_exp)
axs[2].set_title("Exponencial")

axs[3].boxplot(dados_poiss)
axs[3].set_title("Poisson")

axs[4].boxplot(dados_bin)
axs[4].set_title("Binomial")

plt.show()
```

```python
media_teorica = [
    distribuicoes["Normal"]["params"]["loc"],
    distribuicoes["Exponencial"]["params"]["scale"],
    (distribuicoes["Uniforme"]["params"]["low"] + distribuicoes["Uniforme"]["params"]["high"]) / 2,
    distribuicoes["Poisson"]["params"]["lam"],
    distribuicoes["Binomial"]["params"]["p"]
]
```

```python
media_amostra = [
    distribuicoes["Normal"]["estatisticas"]["média"],
    distribuicoes["Exponencial"]["estatisticas"]["média"],
    distribuicoes["Uniforme"]["estatisticas"]["média"],
    distribuicoes["Poisson"]["estatisticas"]["média"],
    distribuicoes["Binomial"]["estatisticas"]["média"]
]
```

```python
nomes = ['Normal','Exponencial','Uniforme','Poisson','Binomial']

fig, axs = plt.subplots(1, 5, figsize=(20,5))

for i in range(5):
    axs[i].bar(["Teórica","Amostral"], [media_teorica[i], media_amostra[i]])
    axs[i].set_title(nomes[i])
    axs[i].set_ylabel("Média")
```
