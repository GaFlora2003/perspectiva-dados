
## 17/03 — Introdução ao Git  

###  Clonar repositório
git clone 
Copia um repositório da nuvem para a sua máquina.

###  Verificar alterações
git status  
Mostra o estado dos arquivos (modificados, adicionados, etc).

git diff  
Mostra exatamente o que foi alterado nos arquivos.

### Enviar alterações para o GitHub
git add
Adiciona um arquivo específico para commit.

git commit -m "mensagem"  
Salva as alterações com uma descrição.

git push  
Envia as alterações para o repositório remoto.

### Boas práticas com Git

Adicionar arquivos específicos:  
git add 

Evitar:  
git add .  

Remover do stage:  
git restore --staged 

Desfazer alterações locais:  
git restore

Histórico de commits:  
git log  


## 24/03 — Gerenciamento com UV  

### O que é UV?

UV é um gerenciador moderno de pacotes Python (substitui pip + venv de forma mais rápida e organizada).


### Inicializar projeto

uv init  
Cria a estrutura inicial do projeto. Usar apenas uma vez.



### Sincronizar dependências

uv sync  
Instala todas as dependências do projeto.


### Instalar pacotes

uv add

Exemplo:  
uv add jupyter pandas numpy  


### Rodar comandos no ambiente

uv run 

Exemplo:  
uv run jupyter notebook  

Se aparecer erro "program not found", significa que o pacote não foi instalado.

### Jupyter

uv add jupyter  
uv run jupyter notebook  


## Manipulação de arquivos no terminal

cat 
Mostra o conteúdo de um arquivo.

## Ignorar arquivos no Git

Criar arquivo:  
touch .gitignore  

Exemplo de conteúdo:  
.venv/  
__pycache__/  
*.pyc  
.env  

Ignora ambiente virtual, arquivos temporários e configurações locais.
