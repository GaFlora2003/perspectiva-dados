```python

df = pd.read_csv('iris.csv')

plt.hist(df['sepal.length'], bins=20, color='steelblue', edgecolor='black')
plt.title('Distribuição do Comprimento da Sépala')
plt.xlabel('Largura da Sépala')
plt.ylabel('Frequência')
plt.show()

cores = {'Setosa': 'red', 'Versicolor': 'green', 'Virginica': 'blue'}

for variedade, grupo in df.groupby('variety'):
    plt.scatter(grupo['sepal.length'], grupo['sepal.width'], label=variedade, color=cores[variedade], alpha=0.7)

plt.xlabel('Comprimento da Sépala')
plt.ylabel('Largura da Sépala')
plt.title('Comprimento vs Largura')
plt.legend()
plt.show()
```

```python
plt.scatter(df['petal.length'], df['petal.width'], s=df['sepal.width'] * 20, alpha=0.6, color='darkorange', edgecolors='black')
plt.xlabel('Comprimento da Pétala')
plt.ylabel('Largura da Pétala')
plt.title('Pétala: Comprimento vs Largura')
plt.show()
```

```python
variedades = df['variety'].unique()
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for i, variedade in enumerate(variedades):
    dados = df[df['variety'] == variedade]['petal.length']
    axes[i].hist(dados, bins=10, color='mediumpurple', edgecolor='black')
    axes[i].set_title(variedade)
    axes[i].set_xlabel('Comprimento da Sépala')
    axes[i].set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
colunas = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
dados_box = [df[col].values for col in colunas]

plt.boxplot(dados_box, labels=['Sep. Comp.', 'Sep. Larg.', 'Pet. Comp.', 'Sep. Larg.'])  # erro
plt.title('Boxplots das Medidas do Iris')
plt.ylabel('Valor (cm)')
plt.show()
```

```python
medias = df.groupby('variety')[colunas].mean()
x = range(len(colunas))
largura = 0.25

fig, ax = plt.subplots(figsize=(10, 5))

for i, variedade in enumerate(medias.index):
    ax.bar([p + largura * i for p in x], medias.loc[variedade], width=largura, label=variedade)

ax.set_xticks([p + largura for p in x])
ax.set_xticklabels(['Sep.Comp', 'Sep.Larg', 'Pet.Comp', 'Pet.Larg'])
ax.set_ylabel('Média (cm)')
ax.set_title('Média das Medidas por Variedade')
ax.legend()
plt.show()
```
