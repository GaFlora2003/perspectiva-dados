```python
matriz_norm = np.random.normal(size=(5,5))
print(matriz_norm)

for i in range(5):
    soma_linhas = sum(matriz_norm[i])
    print('Soma da linha', i+1, soma_linhas)

for j in range(5):
    soma_colunas = sum(matriz_norm[:, j])
    print('Soma da coluna',j+1,soma_colunas)


diag_norm = np.diag(matriz_norm)
print(diag_norm)


```
